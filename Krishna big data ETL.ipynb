```python
from pyspark.sql.functions import col, current_timestamp, to_timestamp, split, hour, when, regexp_replace, lower

# --- 1. BRONZE LAYER (Raw Ingestion) ---
# In Databricks, you would typically read from DBFS or mounted S3/Blob storage.
# Assuming files are uploaded to DBFS (FileStore/tables/...)

# Read raw CSVs
vld_raw = spark.read.csv("/Volumes/workspace/default/krishna_raw_data/VLD_GPS_Data.csv", header=True, inferSchema=True)
etm_raw = spark.read.csv("/Volumes/workspace/default/krishna_raw_data/ETM_Load_Data.csv", header=True, inferSchema=True)
feedback_raw = spark.read.csv("/Volumes/workspace/default/krishna_raw_data/Feedback_Text_Data_Revised.csv", header=True, inferSchema=True)

# Add Ingestion Timestamp (Audit Column)
vld_bronze = vld_raw.withColumn("ingestion_timestamp", current_timestamp())
etm_bronze = etm_raw.withColumn("ingestion_timestamp", current_timestamp())
feedback_bronze = feedback_raw.withColumn("ingestion_timestamp", current_timestamp())

# Write to Bronze Delta Tables
vld_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze_vld")
etm_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze_etm")
feedback_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze_feedback")

print(" Bronze Layer Created: Raw data ingested.")

# --- 2. SILVER LAYER (Cleansed & Enriched) ---

# --- A. Process VLD (GPS) Data ---
df_vld = spark.read.table("bronze_vld")

silver_vld = df_vld.withColumn(
    "Timestamp", to_timestamp(col("Timestamp"), "dd-MM-yyyy HH:mm")
).filter(
    (col("Delay_Minutes") <= 60) & (col("Delay_Minutes") >= -10)  # Remove outliers
).withColumn(
    "Route_ID", split(col("Trip_ID"), "_").getItem(0)  # Extract Route
).withColumn(
    "Hour", hour(col("Timestamp"))  # Extract Hour for joining
)

# --- B. Process ETM (Ticketing) Data ---
df_etm = spark.read.table("bronze_etm")

silver_etm = df_etm.withColumn(
    "Boarding_Time", to_timestamp(col("Boarding_Time"), "dd-MM-yyyy HH:mm")
).withColumn(
    "Route_ID", split(col("Trip_ID"), "_").getItem(0)
).withColumn(
    "Hour", hour(col("Boarding_Time"))
).withColumn(
    "Is_Overcrowded", when(col("Current_Load") > 60, True).otherwise(False)
).withColumn(
    "Is_Underutilized", when(col("Current_Load") < 30, True).otherwise(False)
)

# --- C. Process Feedback Data ---
df_feedback = spark.read.table("bronze_feedback")

silver_feedback = df_feedback.withColumn(
    "Complaint_Time_Approx", to_timestamp(col("Complaint_Time_Approx"), "dd-MM-yyyy HH:mm")
).withColumn(
    "Text_Clean", lower(regexp_replace(col("Text_Content"), "[^a-zA-Z0-9\\s]", "")) # Remove special chars
).withColumn(
    "Link_Status", when(col("Simulated_NER_Bus_ID").isNull(), "Unlinked").otherwise("Linked")
)

# Write to Silver Delta Tables
silver_vld.write.format("delta").mode("overwrite").saveAsTable("silver_vld")
silver_etm.write.format("delta").mode("overwrite").saveAsTable("silver_etm")
silver_feedback.write.format("delta").mode("overwrite").saveAsTable("silver_feedback")

print(" Silver Layer Created: Data cleansed and enriched.")
```
